```python
import urllib.request
import bs4
import matplotlib.pyplot as plt


url ="http://smartlab.tmit.bme.hu/oktatas-deep-learning"


#downloading and formating the text 
plainshit = urllib.request.urlopen(url).read()
soup = bs4.BeautifulSoup(plainshit)

print(soup.get_text())


letters = {} #creating a dict where the keys will be the letters and the values of the keys will be the frequency of the letters

for i in soup.get_text():
  if((i>='a' and i<='z') or (i>='A' and i<='Z')): #checking if the inspected character is part of the alphabet
    if(i not in letters): # if the inspected character cannot be found at the dict, we add it and give it 0 value
      letters[i.lower()] = 1
    else: # if the character already participated at the dict, we add 1 to it's value
      letters[i.lower()]+=1


plt.bar(list(letters.keys()), letters.values(), color = 'g')

print(letters)
```
